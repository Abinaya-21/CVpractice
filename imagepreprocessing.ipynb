##Rotation

```python
(rows, cols) = img.shape[:2]
```

```python
#rotating by 90 degrees
rotate = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
res = cv2.warpAffine(img, rotate, (cols, rows))
plt.imshow(res)
```

##Shifting

```python
shift = np.float32([[1, 0, 100], [0, 1, 50]])
res1 = cv2.warpAffine(img, shift, (cols, rows))
plt.imshow(res1)
```

##Low pass filter (smoothing , blurring)

```python
img3 = cv2.imread('lpf.jpg')
plt.imshow(img3)
```

```python



kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img3,-1,kernel)
plt.imshow(dst)

```

```python
#blurring using averging filter
blur = cv2.blur(img3,(5,5))
plt.imshow(blur)
```

```python
#Gaussian filter(non-uniform) - filter coefficients will not be same
blur1 = cv2.GaussianBlur(img3,(5,5),0)
plt.imshow(blur1)
```

```python
img4 = cv2.imread('saltandpepper.jpg')
plt.imshow(img4)
```

```python
#median filter

median = cv2.medianBlur(img3,5)
plt.imshow(median)
```

```python
#bilateral filtering - preserve edges 
blur = cv2.bilateralFilter(img3,9,75,75)
plt.imshow(blur)
```

##High pass filters

```python


img5 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

# convolute with proper kernels
laplacian = cv2.Laplacian(img5,cv2.CV_64F)
sobelx = cv2.Sobel(img5,cv2.CV_64F,1,0,ksize=5)  
sobely = cv2.Sobel(img5,cv2.CV_64F,0,1,ksize=5)
plt.imshow(laplacian)
```

##Thresholding

```python
img_ = cv2.imread('lpf.jpg')
plt.imshow(img_)
```

```python
#
img_ = cv2.cvtColor(img_, cv2.COLOR_BGR2GRAY)
ret,thresh1 = cv2.threshold(img_, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img_, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img_, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img_, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img_, 120, 255, cv2.THRESH_TOZERO_INV)


cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
   
```

```python
thresh11 = cv2.adaptiveThreshold(img_, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                          cv2.THRESH_BINARY, 199, 5)
  
thresh22 = cv2.adaptiveThreshold(img_, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 199, 5)

cv2_imshow(thresh11)
cv2_imshow(thresh22)
  
```

```python
ret, thresh111 = cv2.threshold(img_, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)      
cv2_imshow(thresh111)   
```
